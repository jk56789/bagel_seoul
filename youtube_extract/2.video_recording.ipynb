```python
import cv2

# 저장된 비디오 파일 경로
video_path = 'C:/Users/user/bagel_seoul/youtube_extract/download/sub_sample/41~62_A-2.mp4'

# 비디오 파일 불러오기
video_capture = cv2.VideoCapture(video_path)

# 웹캠 설정
webcam_capture = cv2.VideoCapture(0)

# 비디오 파일의 너비와 높이 가져오기
frame_width = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))

# 웹캠 프레임 크기 설정
webcam_capture.set(cv2.CAP_PROP_FRAME_WIDTH, frame_width)
webcam_capture.set(cv2.CAP_PROP_FRAME_HEIGHT, frame_height)

fps = video_capture.get(cv2.CAP_PROP_FPS)

# 비디오 저장 설정
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out_combined = cv2.VideoWriter('output_combined.avi', fourcc, fps, (frame_width * 2, frame_height))
out_webcam = cv2.VideoWriter('output_webcam.avi', fourcc, fps, (frame_width, frame_height))

while True:
    # 저장된 비디오 읽기
    ret1, frame1 = video_capture.read()
    # 웹캠에서 프레임 읽기
    ret2, frame2 = webcam_capture.read()
    # 좌우 반전
    if ret2:
        frame2 = cv2.flip(frame2, 1)
    if ret1 and ret2:
        # 좌우로 프레임 합치기
        combined_frame = cv2.hconcat([frame2, frame1])
        
        # 합쳐진 프레임과 웹캠 프레임 보여주기
        cv2.imshow('Combined Video', combined_frame)
        cv2.imshow('Webcam', frame2)
        
        # 비디오 저장
        out_combined.write(combined_frame)
        out_webcam.write(frame2)

    # 'q' 키를 누르면 종료
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# 모든 창 닫기
video_capture.release()
webcam_capture.release()
out_combined.release()
out_webcam.release()
cv2.destroyAllWindows()
```
